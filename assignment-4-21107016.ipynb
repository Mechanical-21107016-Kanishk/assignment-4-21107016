```python
#1
marks = int(input("Enter your marks: ")) # obtain user input

if marks >= 80: # if-elif-else condition to assign grade
  grade = "A"
elif marks >= 60:
  grade = "B"
elif marks >= 50:
  grade = "C"
elif marks >= 45:
  grade = "D"
elif marks >= 25:
  grade = "E"
else:
  grade = "F"

print("Your grade is", grade) # print grade
```

```python
#2
year = int(input("Enter a year: ")) # obtaining user input

if year % 4 == 0: # conditions to assign leap year or not
  if year % 100 == 0 and year % 400 == 0:
    result = "a leap year"
  elif year % 100 == 0 and year % 400 != 0:
    result = "not a leap year"
else:
  result = "not a leap year"

print(year, "is" , result)
```

```python
#3
import random

counter = 1 # intializing counter

# while loop
while counter < 11:
  num1 = random.randint(0, 9) # assigning a random value between 0 and 9
  num2 = random.randint(0, 9)

  userInput = int(input("Question {}: {} × {} = ".format(counter, num1, num2))) # obtaining user input

  if userInput == num1 * num2: # displaying result
    print("Right!")
  else:
    print("Wrong. The answer is", num1 * num2)

  counter = 1 + counter # incrementing counter
```

```python
#4
among5 = int(input("If the candy is divided evenly among 5 people, how many pieces would be left over? "))  # obtaining user input

among6 = int(input("If the candy is divided evenly among 6 people, how many pieces would be left over? ")) 

among7 = int(input("If the candy is divided evenly among 7 people, how many pieces would be left over? "))

for candies in range(0, 200):
  if candies % 5 == among5 and candies % 6 == among6 and candies % 7 == among7:
    print("The number of candies are", candies)
```
